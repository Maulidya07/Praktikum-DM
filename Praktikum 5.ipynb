# **A. Pengecekan Sebelum Dilakukan Cleaning**

Beberapa tahapan sebelum dilakukan cleaning untuk case saat ini
1. **Standarisasi Missing Value** dalam artian menyatukan semua "data kotor" jadi satu format (NaN)
2. **Evaluasi Persentasi Missing Value** untuk menentukan strategi penanganan data (dihapus atau diimputasi) guna mencegah bias, memastikan akurasi analisis statistik, dan menjaga performa model machine learning

```python
df.replace(['UNKNOWN', 'ERROR'], np.nan, inplace =True)
```

# **2. Evaluasi Persentase Missing**

- < 5% Negligible (Dapat diabaikan) = Deletion atau Simple Implutation (Mean / Median)

- 5% - 15% Moderate = Multiple Imputation atau Maximum Likelihood

- 15% -40% Serious/Large = Butuh analisis sensitivitas; hasil mungkin bias

- .> 40% Fatal/Red Flag = Interpretasi sangat beresiko; pertimbangkan membuang variabel tersebut

```python
(df.isnull().sum() / len(df)) * 100
```

# **B. CLEANING**

# **1. Buang kolom location**

karena berbahaya melakukan imputasi dengan tingkat missing value yang tinggi

```python
df.drop(columns=['Location'], inplace=True)
```

# **2. Cleaning kolom item**

```python
df = df.dropna(subset=['Item'])
df['Item'].isnull().sum()
```

# **3. Cleaning Quantity, Price Per Unit**

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Quantity'].plot(kind='box')
```

```python
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Price Per Unit'].plot(kind='box')
```

```python
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)
df['Price Per Unit'].fillna(df['Price Per Unit'].mean(), inplace=True)

df[['Price Per Unit', 'Quantity']].isnull().sum()
```

# **4. Cleaning Total Spent dengan derived feature**

```python
df['Total Spent'] = df['Quantity'] * df['Price Per Unit']
```

# **5. Cleaning Payment Method**

Walaupun memiliki tingkat missing value yang tinggi kenapa tidak langsung di drop saja? karena masih memiliki nilai bisnis yang masih bisa untuk kita analisis

```python
df['Payment Method'].fillna('Unknown', inplace=True)
```

# **6 Cleaning & Formating Transaction Date**

Tipe data time series sangat penting dan juga sulit diimputasi secara akurat

```python
df = df.dropna(subset=['Transaction Date'])
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

# **D. Grouphing & Agregasi**

```python
grouped_byPymentmethod = df.groupby('Payment Method').agg({
    'Total Spent': 'sum',
    'Quantity': 'sum',
    'Transaction ID': 'count'
}).rename(columns={
    'Total Spent': 'Total Spent',
    'Quantity': 'Total Quantity',
    'Transaction ID': 'Total Transactions'
})

grouped_byPymentmethod
```
