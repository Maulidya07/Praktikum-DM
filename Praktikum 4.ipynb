# Filtering Data

```python
# Mengambil data yang hanya di propinsi DKI Jakarta
df_Jakarta = df[df['propinsi'] == 'DKI Jakarta']
df_Jakarta.head()
```

```python
# Mengambil data yang hanya total tempat tidur lebih dari 100
df_rumahSakitBesar = df[df['total_tempat_tidur'] > 100]
df_rumahSakitBesar.head()
```

```python
# Mengambil daya yang ganya di propinsi DKI Jakarta dan total tempat tidur lebih dari 100
df_filter = df[(df['propinsi'] == 'DKI Jakarta') & (df['total_tempat_tidur'] > 100)]
df_filter.head()
```

```python
# Mengambil data yang hanya di propinsi Papua dan total tenaga kerja kurang dari 10
df_filter_RumahSakitKecil = df[(df['propinsi'] == 'Papua') & (df['total_tenaga_kerja'] < 10)]
df_filter_RumahSakitKecil.head()
```

```python
# Mengambil data untuk menampilkan baris dengan kolom kelas bernilai A atau B
df_kelas = df[df['kelas'].isin(['A', 'B'])]
df_kelas.head()
```
