# **Latihan 1**

```python
import math
from collections import Counter

# Data training
data_training = [
    {"temperatur": 10, "angin": 0, "kelas": "Dingin"},
    {"temperatur": 25, "angin": 0, "kelas": "Panas"},
    {"temperatur": 15, "angin": 5, "kelas": "Dingin"},
    {"temperatur": 20, "angin": 3, "kelas": "Panas"},
    {"temperatur": 18, "angin": 7, "kelas": "Dingin"},
    {"temperatur": 20, "angin": 10, "kelas": "Dingin"},
    {"temperatur": 22, "angin": 5, "kelas": "Panas"},
    {"temperatur": 24, "angin": 6, "kelas": "Panas"}
]

# Data testing
temperatur_test = 16
angin_test = 3

# Nilai K
k = 3

# Hitung jarak Euclidean
hasil_jarak = []

for data in data_training:
    jarak = math.sqrt(
        (data["temperatur"] - temperatur_test) ** 2 +
        (data["angin"] - angin_test) ** 2
    )

    hasil_jarak.append({
        "temperatur": data["temperatur"],
        "angin": data["angin"],
        "kelas": data["kelas"],
        "jarak": jarak
    })

# Urutkan berdasarkan jarak terkecil
hasil_jarak.sort(key=lambda x: x["jarak"])

# Ambil K tetangga terdekat
tetangga_terdekat = hasil_jarak[:k]

# Voting kelas
daftar_kelas = []

for tetangga in tetangga_terdekat:
    daftar_kelas.append(tetangga["kelas"])

hasil_voting = Counter(daftar_kelas)
kelas_prediksi = hasil_voting.most_common(1)[0][0]

# Tampilkan hasil
print("=== Hasil Perhitungan Jarak ===")
for data in hasil_jarak:
    print(
        data["temperatur"],
        data["angin"],
        data["kelas"],
        "Jarak:",
        round(data["jarak"], 2)
    )

print("\n=== Tetangga Terdekat ===")
for data in tetangga_terdekat:
    print(data)

print("\nHasil voting:", hasil_voting)
print("Prediksi kelas untuk temperatur 16°C dan angin 3 km/jam adalah:", kelas_prediksi)
```

# **Latihan 2**

```python
# Data hasil sebenarnya dan hasil prediksi
data_kelulusan = [
    {"nim": "T001", "sebenarnya": "Lulus", "prediksi": "Lulus"},
    {"nim": "T002", "sebenarnya": "Lulus", "prediksi": "Lulus"},
    {"nim": "T003", "sebenarnya": "Lulus", "prediksi": "Lulus"},
    {"nim": "T004", "sebenarnya": "Lulus", "prediksi": "Tidak Lulus"},
    {"nim": "T005", "sebenarnya": "Lulus", "prediksi": "Tidak Lulus"},
    {"nim": "T006", "sebenarnya": "Tidak Lulus", "prediksi": "Lulus"},
    {"nim": "T007", "sebenarnya": "Tidak Lulus", "prediksi": "Tidak Lulus"},
    {"nim": "T008", "sebenarnya": "Tidak Lulus", "prediksi": "Tidak Lulus"},
    {"nim": "T009", "sebenarnya": "Tidak Lulus", "prediksi": "Tidak Lulus"},
    {"nim": "T010", "sebenarnya": "Tidak Lulus", "prediksi": "Tidak Lulus"}
]

# Kelas positif = Lulus
TP = 0
TN = 0
FP = 0
FN = 0

for data in data_kelulusan:
    sebenarnya = data["sebenarnya"]
    prediksi = data["prediksi"]

    if sebenarnya == "Lulus" and prediksi == "Lulus":
        TP += 1
    elif sebenarnya == "Tidak Lulus" and prediksi == "Tidak Lulus":
        TN += 1
    elif sebenarnya == "Tidak Lulus" and prediksi == "Lulus":
        FP += 1
    elif sebenarnya == "Lulus" and prediksi == "Tidak Lulus":
        FN += 1

# Hitung accuracy, precision, recall
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)

# Tampilkan confusion matrix
print("=== Confusion Matrix ===")
print("TP:", TP)
print("TN:", TN)
print("FP:", FP)
print("FN:", FN)

print("\nTabel Confusion Matrix")
print("Aktual Lulus & Prediksi Lulus:", TP)
print("Aktual Lulus & Prediksi Tidak Lulus:", FN)
print("Aktual Tidak Lulus & Prediksi Lulus:", FP)
print("Aktual Tidak Lulus & Prediksi Tidak Lulus:", TN)

print("\n=== Hasil Evaluasi ===")
print("Accuracy :", round(accuracy * 100, 2), "%")
print("Precision:", round(precision * 100, 2), "%")
print("Recall   :", round(recall * 100, 2), "%")
```
